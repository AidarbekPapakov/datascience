```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp


file_path = 'zomato.csv'

data = pd.read_csv(file_path)

data[:10]
```

```python
data.columns
```

```python
data.drop(['url', 'address', 'menu_item', 'reviews_list'], axis=1)
```

```python
data.info()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data['rate'].unique()
```

```python
def rate_changer (value):
    if value=='NEW' or value=='-':
        return np.nan
    else:
        value = str(value).split('/')
        value=value[0]
        return float(value)
    
data['rate'] = data['rate'].apply(rate_changer)
data.loc[:10,'rate']
```

```python
data['rate'] = data['rate'].fillna(data['rate'].mean())
data['rate'].isnull().sum()
```

```python
data.info()
```

```python
data = data.dropna()
data.loc[:5]
```

```python
data = data.rename(columns={
    'approx_cost(for two people)': '2plates_price',
    'phone': 'phone number',
    'listed_in(type)': 'Type',
})
data.loc[:5]
```

```python
data['2plates_price'].unique()
```

```python
def remove_commas (value):
    value = str(value)
    if ',' in value:
        value = value.replace(',', '')
        return float(value)
    else:
        return float(value)

data['2plates_price'] = data['2plates_price'].apply(remove_commas)
data['2plates_price'].unique()
```

```python
value = Series(data['rate'].value_counts().index)
value = value.sort_values()
```

```python
frequency = Series(data['rate'].value_counts().sort_index())
```

```python
plt.plot(value.values, frequency.values)
plt.show()
```

```python
rest_types = data['rest_type'].value_counts(ascending=False)
rest_types
```

```python
rest_types_lessthan500 = rest_types[rest_types<500]
rest_types_lessthan500
```

```python
def rest_types_others(value):
    if value in rest_types_lessthan500:
        return 'Others'
    else:
        return value
    
data['rest_type'] = data['rest_type'].apply(rest_types_others)  
data['rest_type'].value_counts()  
```

```python
location = data['location'].value_counts(ascending=False)

location_less_than300 = location[location<300]
location_less_than300
```

```python
def location_other(value):
    if value in location_less_than300:
        return 'others'
    else: 
        return value
    
data['location'] = data['location'].apply(location_other)
```

```python
location_data = DataFrame({
    'location': data['location'].value_counts().index,
    'frequency' : data['location'].value_counts().values
})


```

```python
plt.figure(figsize=(8,5))
ax = sns.barplot(x='location', y='frequency', data=location_data, palette='rainbow')
plt.xticks(rotation = 90)
```
