```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp


plt.style.use('classic')
data = pd.read_csv('gas_prices.csv')
data.loc[0, 'Australia'] = data.loc[:14, 'Australia'].mean()

countries = Series(data.columns.drop('Year'), index=np.arange(10))
fisrt_country = countries[3]
second_country = countries[6]

### Plotting gas prices change over 19 years period in Diff countries 
years = np.arange(1990, 2009).astype(str)
plt.plot(years, data[fisrt_country].values, label = fisrt_country, color=(1, 0.4, 0, 0.8))
plt.plot(years, data[second_country].values, label = second_country, color=(0.15, 0, 1, 0.8))


plt.xticks(rotation = 90)

plt.title('Gas price change over 1990-2008 years', fontdict={'size': 14, 'family': 'fantasy', 'weight': 'bold'})
plt.grid(alpha=0.15)

plt.xlabel('Year', fontdict={'fontsize': 11})
plt.ylabel('Price $/Gallon' , fontdict={'fontsize': 11})
plt.legend()

data

```

```python
def string_lbs_to_num_kg(value):
    value=str(value)
    if 'lbs' in value:
        value = value.replace('lbs','')
    elif value=='nan':
        return 0
    
    value = int(value)
    value = value/2.205
    return value

def string_feet_to_num_metres(value):
    value=str(value)
    if value!='nan':
        value = value.split("'")
        first = float(value[0])
        second = float(value[1])
        second=second/12
        value = first+second
        value = value*30.48
        return value 
    else: 
        return 0    

       


    

heigt = fifa2['Height'].apply(string_feet_to_num_metres)
weight = fifa2['Weight'].apply(string_lbs_to_num_kg)

fig, axes = plt.subplots(2, figsize=(12,9), dpi=100)

axes[0].hist(weight, bins = 50, edgecolor = 'white', color=(0,0.5,1,0.45))
axes[0].set_xlabel('Weigt of Players in kg', fontdict={'weight': 'bold'})
axes[0].set_ylabel('Number of players', fontdict={'weight': 'bold'})

axes[1].hist(heigt, bins = 50, edgecolor = 'white', color=(1,0.4,0,.65))
axes[1].set_xlabel('Height of Players in cm', fontdict={'weight': 'bold'})
axes[1].set_ylabel('Number of Players', fontdict={'weight': 'bold'})
```

```python
height = heigt[heigt!=0]
weight = weight[weight!=0]
height_and_weight = height*weight


height_mean = height.mean()
weight_mean = weight.mean()
height_and_weight_mean = height_and_weight.mean()

height_std = height.std()
weight_std = weight.std()



length=18159

height_and_weight = height_and_weight.reindex(np.arange(18159))

[height_mean, weight_mean, height_std, weight_std, len(height), len(weight)]

corr = (height_and_weight_mean-height_mean*weight_mean)/(height_std*weight_std)
corr

x_points = np.arange(50, 110, 0.2)


plt.style.use('fast')



newfig, newax = plt.subplots(figsize = (8,5), dpi = 120)



newax.set_title('Weight and Height scatter plot', fontdict={'weight': 'bold', 'size': 11})
newax.set_xlabel('Weight')
newax.set_ylabel('Height')
newax.scatter(weight.values, height.values, s=15, c=(0, 0.5, 1, 0.5), linewidths=0)
newax.plot(x_points, corr*(height_std/weight_std)*(x_points-weight_mean)+height_mean, label = 'linear regression equation')
newax.legend()
```
