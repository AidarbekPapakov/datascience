```python
data = pd.read_csv('Engineering_graduate_salary.csv')

df = data
df.columns
df.drop(['ID', 'CollegeID', 'DOB', '10board', '12graduation', '12board', 'CollegeCityID', 'CollegeCityTier', 'GraduationYear'], axis=1, inplace=True)
df.drop('CollegeState', axis=1, inplace=True)

df.drop_duplicates()

```

```python
df.describe()
print(df.loc[df['collegeGPA']<40,'collegeGPA'])

df = df[df['collegeGPA']>40]
# df.reindex(np.arange(2989))[290:301]
df = df.reset_index(drop=True)
df.loc[290:301,:]
df = df.drop(columns='index')
```

```python
salaries_by_degrees = df.groupby('Degree')['Salary'].mean()
```

```python
salaries_by_specialization = df.groupby('Specialization')['Salary'].mean().sort_values()
```

```python
pivot_table_of_salaries = df.pivot_table(values='Salary', index='Specialization', columns='Degree', aggfunc='mean')
```

```python
specializations = df['Specialization'].value_counts()
specializations

specializations_less_than_10 = specializations[specializations<10]
specializations_less_than_10


def clustering_specializations(value): 
    if value in specializations_less_than_10:
        return 'other'
    else:
        return value


df['Specialization'] = df['Specialization'].apply(clustering_specializations)
df['Specialization'].value_counts()
```

```python
plt.figure(figsize=(15,10), dpi=150)
plt.scatter(df.index, df['collegeGPA'], s=5, alpha=0.8)
plt.xlabel("Student's ID", fontdict={'weight': 'bold', 'size': 20})
plt.yticks(np.arange(0, 110, 10))
plt.ylabel("Student's GPA", fontdict={'weight': 'bold', 'size': 20})
plt.grid(alpha=0.2)
```

```python
df.groupby('Gender')['Salary'].quantile(np.arange(0,1.1,0.1))


fig, axes = plt.subplots(2, figsize=(12,8), dpi= 130)
axes[0].hist(df.loc[df['Gender']=='m', 'Salary'], bins=100, edgecolor='white')
axes[1].hist(df.loc[df['Gender']=='f', 'Salary'], bins=100, color='r', edgecolor='white')
```

```python
df = df.replace(-1, np.nan)
null_columns = [column for column in df.columns if df.isna().sum()[column]>0]

for column in null_columns:
    df[column] = df[column].fillna(df[column].mean())
```
